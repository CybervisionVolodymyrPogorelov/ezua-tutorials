# Superset
This chapter will delve into the process of creating a Superset Dashboard using data queried with Presto using HPE Ezmeral Unified Analytics Software Platform.

### 1. Create a Cached Asset
In order to query data head to the Data Engineering, __Query Editor__ section.

This section should look similar to the following picture.


Choose the dataset(s) you want to query


Add a workspace to enter a query
Enter any query and click on __Run__


To view the query result click the arrow left to the Query ID of the result
For saving this query as cached asset click on actions on the very right of your query result and save as __view__

Enter a name and select a schema for your View 

### 2. Create Database in Superset
__Switch to Superset__


Click on the + icon to add a database at the very right 
Select __Presto__  

  
In order to access Cached Assets enter the following __SQL Alchemy URI:__
presto://ezpresto.<domain>:443/cache  
    In this example this results in following URI: presto://ezpresto.hpe-discover-23-ezaf.com:443/cache  
    Provide a name and click on connect

    <b>Note:</b>  If you want to add a datasource directly use the datasource name instead of cache
    </div>


### 3. Create a Dataset in Superset
Head to the __Datasets__ section in Superset   
And click on + DATASET  
Enter the Database you created before as well as the fitting schema and table schema, this could look like this  


Hit add Dataset and create Chart.

### 4. Create Charts in Superset
We can use our created dataset now to create a chart. There are several types of charts available. We are going through two examples to showcase some of the configuration options for different charts.  
First chart we're looking at is the __Bar chart__. This is already shown under Popular. Please keep in mind to always select your dataset on the top when adding a new chart.   
Select the Bar Chart to continue.

We can see that as Chart Source our dataset is already used. The columns available in this data set are shown on the left. On the right we can see a preview of the chart which is currently empty.   

In this view we can still change the type of charts. The Query section appears at every type of chart. The yellow ! in a circle indicates that this field must be filled out for this kind of chart. Which fields are mandatory can differ depending on the selected chart type.

What you want to show in your chart defines how to fill these fields. As we want to create a chart which shows the Total Sales of Products per Year we add as Metric the sum of totalsales. Either you drag and drop from the left into Metrics totalsales or you can click on +. It opens a new window where you can choose between a simple query or choose a custom sql. We are going with the SUM of the possible aggregate options in the simple query.  
  
We don't need a filter. As dimension we choose year, as we want to visualize the Sales of Products per Year. To show it of the products we need porduct as a breakdown.   
  
After selecting this and clicking on create chart you can see a preview of your chart on the right. 

Next to Data you have the option to customize the chart. This includes setting a cholor scheme as well as other things. We want to have an interactive chart where the user can decide between a stacked and a grouped bar chart. Therefore we enable __Extra Controls__.   


Click Update Chart to see the updated version in the preview. Don't forget to save your chart!  
  


As next chart we are gonna create a tree map. Therefore go to charts and add a new one, select your dataset and as tyoe select tree map. With this chart we want to visualize the sold amount of a product per country. Therefore we need to add two Dimensions: product and country. As Metric we are using the SUM of totalsales. Click create chart to get a preview. It should look similar to this.  
  
Don't forget to save your chart.

### 5. Create Dashboard in Superset

When you save a chart Superset asks you if you want to add this chart to an existing Dashboard or create a new one. If you create a dashboard from scratch head to the dashboard section and select +DASHBOARD.

All the charts created by you and also other users are visible now. You can add them with drag and drop and also decide on the place and width it has on your dashboard.   
  

Combine charts as you please. You can also add layout elements to your chart. Don't forget to give it a name.
