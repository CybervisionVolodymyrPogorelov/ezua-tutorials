# Airflow

With Airflow we want to load the data of the CSV files we created and store inside EZUA with 1-Consumer.ipynb into databases and clean the data. Therefore we have 3 DAGs one for each store and therefore one for each Database.

If you want to deploy your own DAGs you have to let the Airflow point to your own Git repo. In order to do that you have to configure Airflow. Click on Configure.



Provide the repo at __repo:__, the branch at __branch:__ and the subFolder at __subDir:__. Make sure you also don't forget to provide a user and their credentials which can be used to access if this is a private repo.

<div class="alert alert-block alert-success">
    <b>Note:</b>  The user has to be maintainer of the repo in order for you to see the DAGs in Airflow UI.
    </div>

In the Airflow UI you should see these 3 DAGs after providing the UI to the Git Repo with the DAGs as well as the user (e.g. https://ddgit.mydirk.de/isteinhauser/end2endairflow.git).

Let's explore one DAG, for example the last one for Germany. When you select the DAG you can see multiple tabs. In the calendar section you can explore when the past runs were. The details section as well as the code section is also interesting, let's head to _Code_ and see how this DAG is defined. 
     
   
A DAG consists of several Tasks. It's coded in Python and in this example our DAG consists of two tasks which are each calling a function. First the _process_csv_ Task is executed which calls the _process_csv_file_ function.
In the DAG you define the schedule interval in which the DAG will be executed.
  

With these DAGs we load and clean the data into their databases. In the next step we can query the data with Presto.
